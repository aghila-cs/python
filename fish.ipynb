```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import StandardScaler,LabelEncoder
from sklearn.metrics import silhouette_score
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df)
plt.xticks(rotation=45)
plt.title('Boxplot for the dataset of fish ')
plt.show()
```

```python
outliers = ['242.0',	'23.2',	'25.4',	'30.0',	'38.4',	'13.4']
for col in outliers:
  Q1 = df[col].quantile(0.25)
  Q3 = df[col].quantile(0.75)
  IQR = Q3 - Q1
  lower_limit = Q1 - 1.5 * IQR
  upper_limit = Q3 + 1.5 * IQR
  df[col] = df[col].clip(lower = lower_limit, upper = upper_limit)
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df)
plt.title('Boxplot for the dataset of fish')
plt.xticks(rotation=45)
plt.show()
```

```python
le = LabelEncoder()
df['Encoded'] = le.fit_transform(df['Bream'])
df_encoded = df.drop(columns = ['Bream'])

# Standard Scaler
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_encoded)
```

```python
df_scaled
```

```python
scaled_df = pd.DataFrame(df_scaled, columns = df_encoded.columns)
```

```python
scaled_df
```

```python
kmeans = KMeans(n_clusters = 3, random_state = 42)
df['KMeans_cluster'] = kmeans.fit_predict(scaled_df)
```

```python
df
```

```python
agg_cluster = AgglomerativeClustering(n_clusters = 3)
df['Agglomerative_Cluster'] = agg_cluster.fit_predict(df_scaled)
```

```python
df
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 2)
df_pca = pca.fit_transform(df_scaled)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 6))
plt.subplot(1, 2, 1)

sns.scatterplot(x=df_pca[:, 0], y=df_pca[:, 1], hue=df['KMeans_cluster'],
                palette=['darkorange', 'teal', 'mediumorchid'], s=100)
plt.title('K-Means Clustering')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.show()
```

```python
plt.figure(figsize = (15,6))
plt.subplot(1, 2, 2)
sns.scatterplot(x = df_pca[:, 0], y = df_pca[:, 1], hue = df['Agglomerative_Cluster'], palette = ['violet', 'brown', 'green'], s = 100)
plt.title('Agglomerative Clustering')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
```
